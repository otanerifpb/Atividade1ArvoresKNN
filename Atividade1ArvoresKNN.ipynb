Passo 01 - Carregar a base TITANIC do seaborn

```python
from seaborn import load_dataset

df = load_dataset('titanic')

df
```

Passo 02 - Observar a relação dos atributos "pclass" com "class"

```python
# Os valores de 'pclass' e 'class', são idênticos, porem um é numérico e outro é categórico, vou eliminar a coluna 'class' por ser categórica
df = df.drop('class', axis = 1)

df
```

Passo 03 -  Observar a quantidade de NaN existente no atributo "deck"

```python
import pandas as pd

df = pd.DataFrame(data = df)

coluna = df['deck']
cont_nan = coluna.isna().sum()

# Devido a quantidade alta de NaN na coluna 'deck', será melhor remover a mesma
df = df.drop('deck', axis = 1)

print()
print('Quantidade de Linhas que contem NaN')
print("Total NaN: " + str(cont_nan))
print()
print('df com a coluna deck removida')
df
```

Passo 04 - Remover as linhas que teham valores NaN

```python
df = df.dropna()

df
```

Passo 05 - Transformar todos os atributos categóricos usando One-Hot Encoder

```python
from sklearn import preprocessing
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = make_column_transformer((OneHotEncoder(), ['sex','embarked','who', 'embark_town','alive', 'alone']), remainder='passthrough')

df = one_hot_encoder.fit_transform(df)
df = pd.DataFrame(data = df)

df
```
